**MERGING :inner,outer,left,right**

```
df1 = pd.DataFrame({"branch":['hyd','blr','vizag','pune'],
                   "total":[90,96,98,92]})
df1 
```

```
df2 = pd.DataFrame({"branch":['blr','hyd','vizag','chennai'],
                   "distance_from_city":[40,60,12,20]})
df2
```

```
pd.merge(df1,df2,on="branch",how="inner",indicator=True)
# INDICATOR FOR FINDING ORIGIN OF ROW
```

```
pd.merge(df1,df2,on="branch",how="outer",indicator=True)
```

```
pd.merge(df1,df2,on="branch",how="left",indicator=True)
```

```
pd.merge(df1,df2,on="branch",how="right",indicator=True)
```

**Arithmatic operations**
1. add() +
2. sub() -
3. mul() *
4. div() /
5. floordiv //
6. pow()    **

```
IPL_2018 = pd.DataFrame({'IPL Team': ['CSK', 'SRH', 'KKR', 'RR', 'MI', 'RCB', 'KXIP', 'DD'],
                         'Matches Played': [16, 17, 16, 15, 14, 14, 14, 14],
                         'Matches Won': [11, 10, 9, 7, 6, 6, 6, 5]}
                       )
IPL_2018.set_index('IPL Team', inplace = True)
IPL_2018
```

```
IPL_2017 = pd.DataFrame({'IPL Team': ['MI', 'RPS', 'KKR', 'SRH', 'KXIP', 'DD', 'GL', 'RCB'],
                         'Matches Played': [17, 16, 16, 15, 14, 14, 14, 14],
                         'Matches Won': [12, 10, 9, 8, 7, 6, 4, 3]}
                       )
IPL_2017.set_index('IPL Team', inplace = True)
IPL_2017
```

```
total = IPL_2017 + IPL_2018
total
```

```
# to  avoid nan we use add(),fill_value to fill nan
total = IPL_2017.add(IPL_2018,fill_value=0)
total
```

```
total = IPL_2017.sub(IPL_2018,fill_value=0)
total
```

```
total = IPL_2017.mul(IPL_2018,fill_value=0)
total
```

```
total = IPL_2017.div(IPL_2018,fill_value=0)
total
```

```
total = IPL_2017.floordiv(IPL_2018,fill_value=0)
total
```

```
total = IPL_2017.mod(IPL_2018,fill_value=0)
total
```

```
total = IPL_2017.pow(IPL_2018,fill_value=0)
total
```

**GROUPBY**

```
data = {"name":['a','b','c','d','e','f'],
       "branch":['hyd','vizag','blr','hyd','vizag','blr'],
       "marks":[90,91,92,93,94,95]}
df = pd.DataFrame(data)
df
```

```
g = df.groupby("branch")
g
```

```
for branch,branch_df in g:
  print(branch)
  print(branch_df)
```

```
g.get_group("hyd")
```

```
g.get_group("hyd")["marks"].sum()
```

```
g.sum()
```

```
g.mean()
```

```
g.describe()
```
